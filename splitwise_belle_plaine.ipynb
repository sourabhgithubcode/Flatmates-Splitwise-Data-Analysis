```python
import pandas as pd
import numpy as np

# Clean the Cost column
def clean_cost(value):
    try:
        # Split concatenated values, convert each to float, and sum
        if isinstance(value, str):
            numbers = [float(x) for x in value.split() if x.replace('.', '', 1).isdigit()]
            return sum(numbers)
        else:
            return float(value)
    except:
        return np.nan  # Replace invalid values with NaN

# Apply the cleaning function to Cost column
df['Cost'] = df['Cost'].apply(clean_cost)

# Drop rows where Cost is still NaN
df = df.dropna(subset=['Cost'])

print("Cleaned Cost column:")
print(df[['Category', 'Cost']].head())
```

```python
# Group expenses by category and sum the costs
category_expenses = df.groupby('Category')['Cost'].sum().reset_index()

# Sort the results by 'Cost' in descending order
category_expenses = category_expenses.sort_values(by='Cost', ascending=False)

# Display sorted results
print("Expenses by Category (Sorted by Cost Descending):")
print(category_expenses)

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(category_expenses['Category'], category_expenses['Cost'])
plt.title('Expenses by Category (Sorted)')
plt.xlabel('Category')
plt.ylabel('Total Cost')
plt.xticks(rotation=45)
plt.show()
```

```python
# List of contributor columns
contributors = ['Sourabh Rodagi', 'shiny parunandi', 'Sudipth Depaul', 
                'Prerna Kosta', 'Nishant', 'Shruthi Gopagoni', 
                'Lingineni Sri Harsha Vardhan', 'Rohith', 'Chintu V', 
                'Shreyansh Pamecha']

# Initialize a dictionary to track owes for each person
owes = {contributor: 0 for contributor in contributors}

# Calculate the amount owed by each participant for each row
for i, row in df.iterrows():
    # Find participants involved in this expense (those who have a non-null or non-zero entry)
    participants = [person for person in contributors if row[person] > 0]
    num_participants = len(participants)
    
    if num_participants > 0:
        # Calculate the share per person for this expense
        share_per_person = row['Cost'] / num_participants
        
        # Update each participant's total owes
        for person in participants:
            owes[person] += share_per_person

# Convert the owes dictionary to a DataFrame
owes_df = pd.DataFrame(list(owes.items()), columns=['Name', 'Total Owes'])

# Sort the results by 'Total Owes' in descending order
owes_df = owes_df.sort_values(by='Total Owes', ascending=False)

# Display the results
print("Top Beneficiaries (Who Owes the Most):")
print(owes_df)
```

```python
# Identify group-level expenses (more than one person sharing)
group_expenses = df[df[contributors].sum(axis=1) > 0]

# Sort the results by 'Cost' in descending order
group_expenses_sorted = group_expenses.sort_values(by='Cost', ascending=False)

# Display the results
print("Group-Level Expenses (Shared by Multiple People, Sorted by Cost Descending):")
print(group_expenses_sorted[['Description', 'Cost', 'Category']])

```

```python
# Calculate Total Paid and Total Owes
total_paid = df[contributors].sum().reset_index()
total_paid.columns = ['Name', 'Total Paid']

total_owes = {name: 0 for name in contributors}
for _, row in df.iterrows():
    participants = [person for person in contributors if row[person] > 0]
    num_participants = len(participants)
    if num_participants > 0:
        share_per_person = row['Cost'] / num_participants
        for person in participants:
            total_owes[person] += share_per_person

total_owes_df = pd.DataFrame(total_owes.items(), columns=['Name', 'Total Owes'])
individual_summary = pd.merge(total_paid, total_owes_df, on='Name', how='outer')
individual_summary['Net Balance'] = individual_summary['Total Paid'] - individual_summary['Total Owes']
```

```python
# Step 1: Split participants into creditors and debtors
creditors = individual_summary[individual_summary['Net Balance'] > 0].copy()
debtors = individual_summary[individual_summary['Net Balance'] < 0].copy()

# Debugging: Check the groups
print("Creditors:")
print(creditors)

print("Debtors:")
print(debtors)

# If no creditors or debtors exist, display a message
if creditors.empty or debtors.empty:
    print("No payment recommendations needed. Either no one owes money or no one needs to be paid.")
else:
    # Convert debtors' Net Balance to absolute values
    debtors['Net Balance'] = debtors['Net Balance'].abs()

    # Step 2: Reconcile balances
    payments = []
    for _, debtor in debtors.iterrows():
        for _, creditor in creditors.iterrows():
            if debtor['Net Balance'] == 0:
                break  # Move to the next debtor if current balance is settled
            amount = min(debtor['Net Balance'], creditor['Net Balance'])
            payments.append({
                'From': debtor['Name'],
                'To': creditor['Name'],
                'Amount': round(amount, 2)
            })
            # Update balances
            debtor['Net Balance'] -= amount
            creditor['Net Balance'] -= amount

    # Step 3: Display Payment Recommendations
    payments_df = pd.DataFrame(payments)
    print("Payment Recommendations to Settle Balances:")
    print(payments_df)
```

```python
import pandas as pd

# Convert 'Date' to a proper datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract the month and year for grouping
df['Month'] = df['Date'].dt.to_period('M')  # Example: '2024-06'

# List of contributors (individuals) and Category column
contributors = ['Sourabh Rodagi', 'shiny parunandi', 'Sudipth Depaul', 
                'Prerna Kosta', 'Nishant', 'Shruthi Gopagoni', 
                'Lingineni Sri Harsha Vardhan', 'Rohith', 'Chintu V', 
                'Shreyansh Pamecha']

# Melt the data to long format to associate each contributor with their contribution
melted_df = df.melt(id_vars=['Month', 'Category'], 
                    value_vars=contributors, 
                    var_name='Person', 
                    value_name='Amount')

# Remove rows where 'Amount' is 0 or NaN
melted_df = melted_df[melted_df['Amount'] > 0]

# Group data by Month, Person, and Category, and calculate the total spent
monthly_spending = melted_df.groupby(['Month', 'Person', 'Category'])['Amount'].sum().reset_index()

# Sort the results for better readability
monthly_spending = monthly_spending.sort_values(by=['Month', 'Person', 'Category'])

# Display the results
print("Monthly Spending by Each Person in Each Category:")
print(monthly_spending)

# Save to CSV for further analysis if needed
monthly_spending.to_csv("monthly_spending_summary.csv", index=False)
print("Monthly spending summary saved to 'monthly_spending_summary.csv'")
```
